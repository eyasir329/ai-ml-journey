## Transforming to df from Inbuilt Data Structure

```python
# list
my_list = [
    ['Alice', 25],
    ['Bob', 30],
    ['Charlie', 28]
]
print(my_list)
```

```python
list_df = pd.DataFrame(my_list)
list_df
```

```python
list_df = pd.DataFrame(my_list, columns=['Name', 'Age'], index=[1, 2, 3])
print(type(list_df))
list_df
```

```python
# tuple

my_tuple = (('Alice', 25), ('Bob', 30), ('Charlie', 28))
print(my_tuple)

tuple_df = pd.DataFrame(my_tuple, columns=['Name', 'Age'])
tuple_df
```

```python
# dictionary

my_dict = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 29, 28]
}
print(my_dict)
dict_df = pd.DataFrame(my_dict)
dict_df
```

```python
# complex data

my_data = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'Paris'},
    {'Name': 'Charlie', 'Age': 28}
]
print(my_data)
```

```python
data_df = pd.DataFrame(my_data)
data_df
```

```python
print(type(data_df))
print(type(data_df['Name']))
```

### Changing values and iteration

```python
df.head(5)
```

```python
# row delation

df.drop(0, inplace=True)
df.head(5)
```

```python
# columns deletion

df.drop('Instructor', axis=1, inplace=True)
df
```

```python
# updation

df.head(5)
```

```python
df.loc[1, 'Python Marks'] = 90
df.loc[1, 'CompletionStatus'] = 'Completed'

df.head()
```

```python
# mutiple updation

df.loc[1:3, 'Python Marks'] += 2
df.head()
```

```python
# iteration

for i, series in df.iterrows():
    print(f"{i} : {series}")
    if i == 5:
        break
```

```python
for i in df.itertuples(index=False):
    print(i)
```

## Filtering data based on condition

```python
df
```

```python
# conditional copy

not_started = df.loc[df['CompletionStatus'] == 'Not Started']
not_started
```

```python
completed = df.loc[df['CompletionStatus'] == 'Completed']
completed
```

```python
# completed and DS marks is greater or equal 90

completed_ds90 = df.loc[(df['CompletionStatus'] == 'Completed') & (df['Data Structure Marks'] >= 90)]
completed_ds90
```
